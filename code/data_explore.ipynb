```python
df = df.drop(['customerID'], axis = 1) 
df.head()
```

```python
print("\nNumber of NaN by columns:")
print(df.isnull().sum())
```

```python
#dữ liệu thiếu nma vì nó là một dấu cách [ ] nên không được ghi nhận
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

```python
#in ra 11 cái total charges bị miss
df[np.isnan(df['TotalCharges'])]
```

```python
#miss total charges thì tenure của nó bằng 0
#chỉ xảy ra với tenure của 11 cái này thôi, điểm danh bằng index như dưới
df[df['tenure'] == 0].index
#tenure = 0 tức là chưa gắn bó tháng nào, nhưng cột tiền nong hàng tháng lại không bị trống
```

```python
#drop hết những row vô lí, so với size dataset thì cũng không bõ
df.drop(labels=df[df['tenure'] == 0].index, axis=0, inplace=True)
df[df['tenure'] == 0].index
```

```python
#check lại sau khi đã handle missing
df.isnull().sum()
```

```python
#khá nhiều cột yes no nhưng chỉ mình cột senior citizen là dạng nhị phân
#encode
df["SeniorCitizen"]= df["SeniorCitizen"].map({0: "No", 1: "Yes"})
df.head()
```

```python
#lưu dataset mới vào để sang file visualization
df.to_csv("../dataset/cleaned_dataset.csv", index=False)
```

```python
#biến mấy cái y n thành 0 1
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series

df = df.apply(lambda x: object_to_int(x))
df.head()
```

```python
#chọn x y
X = df.drop(columns = ['Churn'])
y = df['Churn'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 40, stratify=y)
```

```python
#ghép x train và ytrain thành một dataframe
train_df = X_train.copy()
train_df['Churn'] = y_train

test_df = X_test.copy()
test_df['Churn'] = y_test

#lưu file
train_df.to_csv("../dataset/train.csv", index=False)
test_df.to_csv("../dataset/test.csv", index=False)
```
